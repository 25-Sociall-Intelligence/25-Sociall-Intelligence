### 1. 상권별 매출 분석

```python
# 상권별 데이터 개수 확인
## 상권_코드_명별로 데이터를 그룹화 (각 상권이 데이터에 몇번 등장하는지)

area_count = df.groupby('상권_코드_명').size().reset_index(name='데이터_개수')
print("상권별 데이터 수:")
print(area_count.sort_values('데이터_개수', ascending=False).head(20))
```

상권별 데이터 수 기준 상위 10개 상권 : 명동, 종로, 가산디지털단지, 역삼, 노원, 장안동사거리, 강남, 군자, 포이사거리, 논현역 

```python
# 상권별 업종 수 계산
## 각 상권에 몇개의 서로 다른 업종이 있는지 (ex, 강남역에 한식음식점, 미용실 있으면 업종_수 = 2)

area_industry_count = df.groupby('상권_코드_명')['서비스_업종_코드_명'].nunique().reset_index(name='업종_수')

print(area_industry_count.sort_values('업종_수', ascending=False).head(20))

# 명동이 서로 다른 업종이 제일 많음
```

상권별 서비스 업종 수 기준 상위 10개 상권: 명동, 종로, 가산디지털단지, 역삼, 노원, 여의도, 강남, 포이사거리, 장안동사거리, 장지역

```python
# 상권 데이터 개수와 업종 수 사이의 상관관계 분석 

# 두 데이터프레임 병합
area_correlation_data = pd.merge(area_count, area_industry_count, on='상권_코드_명')
print("\n병합된 데이터 샘플:")
print(area_correlation_data.head(10))

# 상관계수 계산
correlation_coef = area_correlation_data['데이터_개수'].corr(area_correlation_data['업종_수'])
print(f"\n데이터 개수와 업종 수 간의 상관계수: {correlation_coef:.4f}")
```

=> 강한 양의 상관관계: 데이터 개수가 많을수록 업종 수도 많음

```python
# 상권별 총 매출액 계산
## 각 상권의 총 당월 매출 금액과 당월 매출 건수 

area_sales = df.groupby(['상권_코드_명', '상권_구분_코드_명']).agg({
    '당월_매출_금액': 'sum',
    '당월_매출_건수': 'sum'
}).reset_index()

print(area_sales)
```

```python
# 데이터 병합
## 상권_코드_명을 기준으로 병합
area_stats = pd.merge(area_sales, area_count, on='상권_코드_명')
area_stats = pd.merge(area_stats, area_industry_count, on='상권_코드_명')
print(area_stats)
```

```python
# 데이터에 기록된 각 항목당 평균 매출액

area_stats['레코드당_평균_매출액'] = area_stats['당월_매출_금액'] / area_stats['데이터_개수']


# 각 업종이 평균적으로 기여하는 매출액

area_stats['업종당_평균_매출액'] = area_stats['당월_매출_금액'] / area_stats['업종_수']


# 고객 한명이 한번 방문할때 지출하는 평균 금액

area_stats['객단가'] = area_stats['당월_매출_금액'] / area_stats['당월_매출_건수']

print(area_stats)
```

```python
# 1-1. 매출액 및 객단가 기준 상위 10개 상권

## 총 매출액 기준 상위 10개 상권 (당월 매출 금액을 기준으로 내림차순으로 정렬) 
top_total_sales = area_stats.sort_values('당월_매출_금액', ascending=False).head(10)
print(top_total_sales)

## 레코드당 평균 매출액 기준 상위 10개 상권
top_avg_per_record = area_stats.sort_values('레코드당_평균_매출액', ascending=False).head(10)
print(top_avg_per_record)

## 업종당 평균 매출액 기준 상위 10개 상권
top_avg_per_industry = area_stats.sort_values('업종당_평균_매출액', ascending=False).head(10)
print(top_avg_per_industry)

## 객단가 기준 상위 10개 상권
top_per_transaction = area_stats.sort_values('객단가', ascending=False).head(10)
print(top_per_transaction)

# 시각화: 4가지 기준으로 비교
fig, axes = plt.subplots(2, 2, figsize=(20, 15))

# 총 매출액
sns.barplot(x='당월_매출_금액', y='상권_코드_명', data=top_total_sales, ax=axes[0,0])
axes[0,0].set_title('총 매출액 기준 상위 10개 상권')
axes[0,0].set_xlabel('총 매출액')
axes[0,0].set_ylabel('상권명')

# 레코드당 평균 매출액
sns.barplot(x='레코드당_평균_매출액', y='상권_코드_명', data=top_avg_per_record, ax=axes[0,1])
axes[0,1].set_title('레코드당 평균 매출액 기준 상위 10개 상권')
axes[0,1].set_xlabel('레코드당 평균 매출액')
axes[0,1].set_ylabel('상권명')

# 업종당 평균 매출액
sns.barplot(x='업종당_평균_매출액', y='상권_코드_명', data=top_avg_per_industry, ax=axes[1,0])
axes[1,0].set_title('업종당 평균 매출액 기준 상위 10개 상권')
axes[1,0].set_xlabel('업종당 평균 매출액')
axes[1,0].set_ylabel('상권명')

# 객단가
sns.barplot(x='객단가', y='상권_코드_명', data=top_per_transaction, ax=axes[1,1])
axes[1,1].set_title('객단가 기준 상위 10개 상권')
axes[1,1].set_xlabel('객단가 (원/건)')
axes[1,1].set_ylabel('상권명')

plt.tight_layout()
plt.show()
```

##### 총 매출액 기준 상위 10개: 노량진역, 가산디지털단지, 용산전자상가, 명동, 가락시장, 강남역, 잠실 관광특구, 종로3가역, 잠실역, 종로청계관광특구
##### 레코드당 평균 매출액 기준 상위 10개: 동서시장, 남정초등학교, 가락시장, 독산동 우시장, 노량진역, 청량리종합시장, 용산전자상가, 청량리수산시장, 강서농산물도매시장, 청량리청과물시장
##### 업종당 평균 매출액 기준 상위 10개: 동서시장, 남정초등학교, 가락시장, 독산동 우시장, 노량진역, 청량리종합시장,  용산전자상가, 강서농산물도매시장, 청량리수산시장, 청량리청과물시장
##### 객단가 기준 상위 10개: 당산역 13번, 아차산역 5번, 길상사, 신우어린이공원, 신내IC북측, 충정로역 7번, 청담초등학교, 한인가구단지, 새석관시장, 한림대학교한강성심병원

```python
# 1-2. 상관관계 분석: 데이터 개수와 총 매출액의 관계
## 데이터 개수 / 업종 수가 많으면 총 매출액도 높은가?

plt.figure(figsize=(12, 5))

# 데이터 개수 와 총 매출액
plt.subplot(1, 2, 1)
plt.scatter(area_stats['데이터_개수'], area_stats['당월_매출_금액'], alpha=0.6)
plt.xlabel('데이터 레코드 수')
plt.ylabel('총 매출액')
plt.title('데이터 개수와 총 매출액')
plt.grid(True, alpha=0.3)

# 상관계수 계산
corr_coef = area_stats['데이터_개수'].corr(area_stats['당월_매출_금액'])
plt.text(0.05, 0.95, f'상관계수: {corr_coef:.3f}', transform=plt.gca().transAxes, 
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# 업종 수와 총 매출액
plt.subplot(1, 2, 2)
plt.scatter(area_stats['업종_수'], area_stats['당월_매출_금액'], alpha=0.6, color='orange')
plt.xlabel('업종 수')
plt.ylabel('총 매출액')
plt.title('업종 수와 총 매출액')
plt.grid(True, alpha=0.3)

# 상관계수 계산
corr_coef2 = area_stats['업종_수'].corr(area_stats['당월_매출_금액'])
plt.text(0.05, 0.95, f'상관계수: {corr_coef2:.3f}', transform=plt.gca().transAxes, 
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()
```

##### 하나의 상권에서 데이터 개수 / 업종수가 많으면 총 매출액이 높아지는 경향은 있으나 항상 그런건 아님
##### 규모가 거의 모든 것을 결정하는 것은 아니다. 

```python
# 1-3. 상권별 업종 매출 분석
area_category_sales = df.groupby(['상권_코드_명', '서비스_업종_코드_명']).agg({
    '당월_매출_금액': 'sum'
}).reset_index()

# 매출액 상위 3개 상권 선택
top3_areas = top_total_sales['상권_코드_명'].head(3).tolist()

# 상위 3개 상권의 업종별 매출 비교
top3_area_category = area_category_sales[area_category_sales['상권_코드_명'].isin(top3_areas)]

# 각 상권별로 상위 5개 업종만 선택
top_categories_by_area = []
for area in top3_areas:
    area_data = top3_area_category[top3_area_category['상권_코드_명'] == area]
    top5 = area_data.sort_values('당월_매출_금액', ascending=False).head(5)
    top_categories_by_area.append(top5)

top_categories_combined = pd.concat(top_categories_by_area)

# 시각화: 상위 3개 상권의 업종별 매출 비교
plt.figure(figsize=(14, 8))
sns.barplot(x='당월_매출_금액', y='서비스_업종_코드_명', hue='상권_코드_명', data=top_categories_combined)
plt.title('상위 3개 상권의 업종별 매출 비교')
plt.xlabel('매출액')
plt.ylabel('업종')
plt.legend(title='상권명')
plt.tight_layout()
plt.show()  
```

##### 노량진 - 수산물 판매, 가산디지털단지 - 일반의류, 용산전자상가 - 컴퓨터 및 주변 장치 판매 업종이 매출이 제일 높음

```python
# 1-4. 상권 유형별 매출 분석 
# 상권 유형별 상권 수 계산
## 고유한 상권의 개수를 세는것, 같은 상권이 여러번 나타나도 1개로 카운트 (ex, 골목상권-강남-한식전문, 골목상권-강남-커피전문, 골목상권-홍대-한식전문점 이면 2개)
area_type_count = df.groupby('상권_구분_코드_명')['상권_코드_명'].nunique().reset_index(name='상권_수')


# 상권 유형별 총 매출액 및 매출 건수
area_type_sales = df.groupby('상권_구분_코드_명').agg({
    '당월_매출_금액': 'sum',
    '당월_매출_건수': 'sum'
}).reset_index()

# 상권 유형별 통계 데이터 병합
area_type_stats = pd.merge(area_type_sales, area_type_count, on='상권_구분_코드_명')

print(area_type_stats)

# 상권 유형별 상권당 평균 매출액 계산
area_type_stats['상권당_평균_매출액'] = area_type_stats['당월_매출_금액'] / area_type_stats['상권_수']

# 상권 유형별 객단가 계산
area_type_stats['객단가'] = area_type_stats['당월_매출_금액'] / area_type_stats['당월_매출_건수']

# 시각화 1: 상권 유형별 총 매출액
plt.figure(figsize=(10, 6))
sns.barplot(x='상권_구분_코드_명', y='당월_매출_금액', data=area_type_stats)
plt.title('상권 유형별 총 매출액')
plt.xlabel('상권 유형')
plt.ylabel('총 매출액')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 시각화 2: 상권 유형별 상권당 평균 매출액
plt.figure(figsize=(10, 6))
sns.barplot(x='상권_구분_코드_명', y='상권당_평균_매출액', data=area_type_stats)
plt.title('상권 유형별 상권당 평균 매출액')
plt.xlabel('상권 유형')
plt.ylabel('상권당 평균 매출액')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 시각화 3: 상권 유형별 객단가
plt.figure(figsize=(10, 6))
sns.barplot(x='상권_구분_코드_명', y='객단가', data=area_type_stats)
plt.title('상권 유형별 평균 객단가')
plt.xlabel('상권 유형')
plt.ylabel('객단가 (원/건)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 시각화 4: 상권 유형별 상권 수와 매출액 비교
fig, ax1 = plt.subplots(figsize=(12, 6))

# 상권 수 
x = np.arange(len(area_type_stats))
ax1.bar(x - 0.2, area_type_stats['상권_수'], width=0.4, color='skyblue', label='상권 수')
ax1.set_ylabel('상권 수', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# 상권당 평균 매출액 
ax2 = ax1.twinx()
ax2.bar(x + 0.2, area_type_stats['상권당_평균_매출액'], width=0.4, color='salmon', label='상권당 평균 매출액')
ax2.set_ylabel('상권당 평균 매출액', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.xticks(x, area_type_stats['상권_구분_코드_명'], rotation=45)
plt.title('상권 유형별 상권 수와 상권당 평균 매출액 비교')
plt.tight_layout()

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.show()
```

##### 발달 상권 유형이 전체 시장에서 큰 비중을 차지하는 것을 확인할 수 있다.
##### 상권 수의 영향을 제외하고 보았을때 관광 특구 유형이 상권당 평균 매출액이 높을것을 확인할 수 있다
##### 발달 상권 유형이 객단가가 제일 높을 것을 확인 할 수 있다.
##### 골목상권이 제일 상권수가 많은 것을 확인할 수 있으나 상권당 평균 매출액은 제일 작다
##### 관광특구 상권은 상권수는 적지만 상권당 평균 매출액이 높을 것을 확인 할 수 있다. (효율성 좋음)

```python
# 1-5. 상권 유형별 서비스 업종 분포 분석

# 1. 상권 유형별 업종 개수 분석
area_industry_count = df.groupby(['상권_구분_코드_명', '서비스_업종_코드_명']).size().reset_index(name='업종_건수')

# 2. 각 상권 유형별 상위 10개 업종 분석
print("=== 상권 유형별 주요 서비스 업종 ===")
for area_type in area_industry_count['상권_구분_코드_명'].unique():
    area_data = area_industry_count[area_industry_count['상권_구분_코드_명'] == area_type]
    top_industries = area_data.sort_values('업종_건수', ascending=False).head(10)
    
    print(f"\n📍 {area_type} 상위 10개 업종:")
    for idx, row in top_industries.iterrows():
        print(f"  {row['서비스_업종_코드_명']}: {row['업종_건수']}개")

# 3. 시각화: 각 상권 유형별 상위 5개 업종
fig, axes = plt.subplots(2, 2, figsize=(20, 15))
area_types = area_industry_count['상권_구분_코드_명'].unique()

for i, area_type in enumerate(area_types):
    ax = axes[i//2, i%2]
    area_data = area_industry_count[area_industry_count['상권_구분_코드_명'] == area_type]
    top5 = area_data.sort_values('업종_건수', ascending=False).head(5)
    
    sns.barplot(x='업종_건수', y='서비스_업종_코드_명', data=top5, ax=ax)
    ax.set_title(f'{area_type} 상위 5개 업종')
    ax.set_xlabel('업종 건수')
    ax.set_ylabel('서비스 업종')

plt.tight_layout()
plt.show()

# 4. 상권 유형별 업종 다양성 분석
area_diversity = df.groupby('상권_구분_코드_명')['서비스_업종_코드_명'].nunique().reset_index(name='업종_다양성')
area_diversity = area_diversity.sort_values('업종_다양성', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='상권_구분_코드_명', y='업종_다양성', data=area_diversity)
plt.title('상권 유형별 업종 다양성 (서로 다른 업종 수)')
plt.xlabel('상권 유형')
plt.ylabel('업종 수')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\n=== 상권 유형별 업종 다양성 ===")
for _, row in area_diversity.iterrows():
    print(f"{row['상권_구분_코드_명']}: {row['업종_다양성']}개 업종")

# 5. 상권 유형별 업종 매출 기여도 분석
area_industry_sales = df.groupby(['상권_구분_코드_명', '서비스_업종_코드_명'])['당월_매출_금액'].sum().reset_index()

print("\n=== 상권 유형별 매출 기여도 상위 5개 업종 ===")
for area_type in area_industry_sales['상권_구분_코드_명'].unique():
    area_data = area_industry_sales[area_industry_sales['상권_구분_코드_명'] == area_type]
    top_sales = area_data.sort_values('당월_매출_금액', ascending=False).head(5)
    
    print(f"\n {area_type} 매출 기여도 상위 5개:")
    for idx, row in top_sales.iterrows():
        sales_billion = row['당월_매출_금액'] / 1e8  # 억 단위 변환
        print(f"  {row['서비스_업종_코드_명']}: {sales_billion:.0f}억원")

# 6. 히트맵: 상권 유형별 주요 업종 분포
# 상위 20개 업종만 선택
top_industries_overall = df['서비스_업종_코드_명'].value_counts().head(20).index
area_industry_matrix = df[df['서비스_업종_코드_명'].isin(top_industries_overall)].groupby(['상권_구분_코드_명', '서비스_업종_코드_명']).size().unstack(fill_value=0)

plt.figure(figsize=(15, 8))
sns.heatmap(area_industry_matrix, annot=True, fmt='d', cmap='YlOrRd')
plt.title('상권 유형별 주요 업종 분포 히트맵')
plt.xlabel('서비스 업종')
plt.ylabel('상권 유형')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 2. 시간대별 매출 패턴 분석

```python
# 2-1. 요일별 매출 분석

weekday_cols = ['월요일_매출_금액', '화요일_매출_금액', '수요일_매출_금액', 
                '목요일_매출_금액', '금요일_매출_금액', '토요일_매출_금액', '일요일_매출_금액']

weekday_sales = df[weekday_cols].sum()
weekday_sales.index = ['월', '화', '수', '목', '금', '토', '일']

# 시각화: 요일별 매출 추이
plt.figure(figsize=(10, 6))
weekday_sales.plot(kind='bar', color='skyblue')
plt.title('요일별 총 매출액')
plt.xlabel('요일')
plt.ylabel('매출액')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()  

## 금요일이 제일 매출이 높음
```

```python
# 2-2. 요일별 객단가 분석
weekday_transactions = ['월요일_매출_건수', '화요일_매출_건수', '수요일_매출_건수', 
                       '목요일_매출_건수', '금요일_매출_건수', '토요일_매출_건수', '일요일_매출_건수']
weekday_amount = weekday_cols
weekday_efficiency = pd.DataFrame()

for i, day in enumerate(['월', '화', '수', '목', '금', '토', '일']):
    weekday_efficiency[day] = [df[weekday_amount[i]].sum() / df[weekday_transactions[i]].sum()]

# 시각화: 요일별 객단가
plt.figure(figsize=(10, 6))
weekday_efficiency.T.plot(kind='bar', legend=False, color='teal')
plt.title('요일별 평균 객단가')
plt.xlabel('요일')
plt.ylabel('객단가 (원/건)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

## 요일별 평균 객단가는 금요일, 토요일이 높다
```

```python
# 2-3. 시간대별 매출 분석

time_cols = ['시간대_00~06_매출_금액', '시간대_06~11_매출_금액', '시간대_11~14_매출_금액',
            '시간대_14~17_매출_금액', '시간대_17~21_매출_금액', '시간대_21~24_매출_금액']

time_sales = df[time_cols].sum()
time_sales.index = ['00~06', '06~11', '11~14', '14~17', '17~21', '21~24']

# 시각화: 시간대별 매출 추이
plt.figure(figsize=(10, 6))
time_sales.plot(kind='bar', color='salmon')
plt.title('시간대별 총 매출액')
plt.xlabel('시간대')
plt.ylabel('매출액')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show() 

## 17 ~ 21 사이가 매출액이 제일 높음
```

```python
# 2-4. 상위 상권들의 시간대별 매출 패턴 분석

# 상위 5개 상권만 선택
top5_areas = top_total_sales['상권_코드_명'].head(5).tolist()
top_areas_df = df[df['상권_코드_명'].isin(top5_areas)]

# 각 상권별 시간대 매출 계산
area_time_sales = pd.DataFrame()

for area in top5_areas:
    area_data = top_areas_df[top_areas_df['상권_코드_명'] == area]
    area_time_data = area_data[time_cols].sum()
    area_time_sales[area] = area_time_data

area_time_sales.index = ['00~06', '06~11', '11~14', '14~17', '17~21', '21~24']

# 시각화: 상위 상권별 시간대 매출 패턴
plt.figure(figsize=(12, 8))
area_time_sales.plot(kind='bar')
plt.title('상위 5개 상권의 시간대별 매출 패턴')
plt.xlabel('시간대')
plt.ylabel('매출액')
plt.legend(title='상권명')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()  

# 노량진 - 17 ~ 21, 가산디지털단지 - 14 ~ 17, 용산전자상가 - 14~17, 명동 - 11 ~ 14 , 가락시장 - 00 ~ 06 매출이 높음
```

```python
# 2-5. 히트맵: 상위 상권의 요일-시간대별 매출 패턴 (첫 번째 상권 기준)

top_area = top5_areas[0]
top_area_df = df[df['상권_코드_명'] == top_area]

weekday_time_data = []
for i, day in enumerate(['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']):
    day_data = []
    for time in ['00~06', '06~11', '11~14', '14~17', '17~21', '21~24']:
        # 해당 데이터가 없는 경우 추정
        if f'{day}_시간대_{time}_매출_금액' in top_area_df.columns:
            value = top_area_df[f'{day}_시간대_{time}_매출_금액'].sum()
        else:
            # 해당 컬럼이 없으면 전체 요일 매출과 시간대 매출의 비율로 추정
            day_total = top_area_df[f'{day}_매출_금액'].sum()
            time_ratio = top_area_df[f'시간대_{time}_매출_금액'].sum() / top_area_df['당월_매출_금액'].sum()
            value = day_total * time_ratio
        day_data.append(value)
    weekday_time_data.append(day_data)

# 히트맵 데이터 생성
heatmap_df = pd.DataFrame(weekday_time_data, 
                          index=['월', '화', '수', '목', '금', '토', '일'],
                          columns=['00~06', '06~11', '11~14', '14~17', '17~21', '21~24'])

# 시각화: 요일-시간대별 매출 히트맵
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_df, annot=True, cmap='YlGnBu', fmt='.0f')
plt.title(f'{top_area} 상권의 요일-시간대별 매출 패턴')
plt.xlabel('시간대')
plt.ylabel('요일')
plt.tight_layout()
plt.show()

## 노량진에서는 토요일 17 ~21 가 제일 매출이 높음 
```

```python
# 2-6. 시간대별 매출과 전체 매출의 상관관계

# 시간대별 매출 컬럼
time_sales_cols = ['시간대_00~06_매출_금액', '시간대_06~11_매출_금액', '시간대_11~14_매출_금액',
                   '시간대_14~17_매출_금액', '시간대_17~21_매출_금액', '시간대_21~24_매출_금액']

# 상관관계 매트릭스
time_corr = df[time_sales_cols + ['당월_매출_금액']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(time_corr, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('시간대별 매출과 전체 매출의 상관관계')
plt.tight_layout()
plt.show()

# 가장 영향력 있는 시간대 찾기
time_correlations = time_corr['당월_매출_금액'][:-1].sort_values(ascending=False)
print("=== 전체 매출에 가장 영향을 주는 시간대 ===")
for time_period, corr in time_correlations.items():
    time_label = time_period.replace('시간대_', '').replace('_매출_금액', '')
    print(f"{time_label}: 상관계수 {corr:.3f}")
```

### 3. 고객 특성별 매출 분석

```python
# 3-2. 연령대별 매출 분석 

# 연령대별 매출액 데이터
age_cols = ['연령대_10_매출_금액', '연령대_20_매출_금액', '연령대_30_매출_금액', 
            '연령대_40_매출_금액', '연령대_50_매출_금액', '연령대_60_이상_매출_금액']

age_sales = df[age_cols].sum()
age_sales.index = ['10대', '20대', '30대', '40대', '50대', '60대 이상']

# 연령대별 매출 건수 데이터
age_transaction_cols = ['연령대_10_매출_건수', '연령대_20_매출_건수', '연령대_30_매출_건수', 
                       '연령대_40_매출_건수', '연령대_50_매출_건수', '연령대_60_이상_매출_건수']

age_transactions = df[age_transaction_cols].sum()
age_transactions.index = ['10대', '20대', '30대', '40대', '50대', '60대 이상']

# 연령대별 객단가 계산
age_avg_price = age_sales / age_transactions

print("=== 연령대별 매출 분석 요약 ===")
for age in age_sales.index:
    print(f"{age}: 매출액 {age_sales[age]:,}원, 매출건수 {age_transactions[age]:,}건, 객단가 {age_avg_price[age]:,.0f}원")

print(f"\n전체 합계: 매출액 {age_sales.sum():,}원, 매출건수 {age_transactions.sum():,}건")


fig, axes = plt.subplots(2, 2, figsize=(16, 12))

## 연령대별 매출액
axes[0, 0].bar(age_sales.index, age_sales.values, color='purple', alpha=0.7)
axes[0, 0].set_title('연령대별 총 매출액')
axes[0, 0].set_xlabel('연령대')
axes[0, 0].set_ylabel('매출액')
axes[0, 0].grid(axis='y', linestyle='--', alpha=0.3)
axes[0, 0].tick_params(axis='x', rotation=45)

##연령대별 매출 건수
axes[0, 1].bar(age_transactions.index, age_transactions.values, color='orange', alpha=0.7)
axes[0, 1].set_title('연령대별 총 매출 건수')
axes[0, 1].set_xlabel('연령대')
axes[0, 1].set_ylabel('매출 건수')
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.3)
axes[0, 1].tick_params(axis='x', rotation=45)

## 연령대별 객단가
axes[1, 0].bar(age_avg_price.index, age_avg_price.values, color='green', alpha=0.7)
axes[1, 0].set_title('연령대별 평균 객단가')
axes[1, 0].set_xlabel('연령대')
axes[1, 0].set_ylabel('객단가 (원)')
axes[1, 0].grid(axis='y', linestyle='--', alpha=0.3)
axes[1, 0].tick_params(axis='x', rotation=45)

## 매출액 vs 매출건수 비교 (정규화)
x = range(len(age_sales))
width = 0.35

# 정규화 (0-1 스케일)
sales_normalized = age_sales / age_sales.max()
transactions_normalized = age_transactions / age_transactions.max()

bars1 = axes[1, 1].bar([i - width/2 for i in x], sales_normalized, width, 
                       label='매출액 (정규화)', color='purple', alpha=0.7)
bars2 = axes[1, 1].bar([i + width/2 for i in x], transactions_normalized, width,
                       label='매출건수 (정규화)', color='orange', alpha=0.7)

axes[1, 1].set_title('연령대별 매출액 vs 매출건수 비교 (정규화)')
axes[1, 1].set_ylabel('정규화된 값 (0-1)')
axes[1, 1].set_xticks(x)
axes[1, 1].set_xticklabels(age_sales.index, rotation=45)
axes[1, 1].legend()
axes[1, 1].grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()

# 연령대별 소비 패턴 인사이트
print("\n=== 연령대별 소비 패턴 ===")

# 최고 매출액/매출건수/객단가 연령대 찾기
max_sales_age = age_sales.idxmax()
max_transactions_age = age_transactions.idxmax()
max_avg_price_age = age_avg_price.idxmax()

print(f"최고 매출액: {max_sales_age} ({age_sales.max():,}원)")
print(f"최고 매출건수: {max_transactions_age} ({age_transactions.max():,}건)")
print(f"최고 객단가: {max_avg_price_age} ({age_avg_price.max():,.0f}원)")
```

```python
# 3-3. 연령대별 서비스 업종 소비 패턴 분석

# 1. 연령대별 업종별 매출액 분석
age_industry_sales = df.groupby(['서비스_업종_코드_명']).agg({
    '연령대_10_매출_금액': 'sum',
    '연령대_20_매출_금액': 'sum', 
    '연령대_30_매출_금액': 'sum',
    '연령대_40_매출_금액': 'sum',
    '연령대_50_매출_금액': 'sum',
    '연령대_60_이상_매출_금액': 'sum'
}).reset_index()

# 2. 각 연령대별 상위 10개 업종 분석
age_columns = ['연령대_10_매출_금액', '연령대_20_매출_금액', '연령대_30_매출_금액',
               '연령대_40_매출_금액', '연령대_50_매출_금액', '연령대_60_이상_매출_금액']

age_labels = ['10대', '20대', '30대', '40대', '50대', '60대+']

print("=== 연령대별 주요 소비 업종 ===")
for i, (col, label) in enumerate(zip(age_columns, age_labels)):
    top_industries = age_industry_sales.sort_values(col, ascending=False).head(10)
    print(f"\n {label} 매출액 상위 10개 업종:")
    for idx, row in top_industries.iterrows():
        sales_billion = row[col] / 1e8  # 억 단위 변환
        print(f"  {row['서비스_업종_코드_명']}: {sales_billion:.0f}억원")

# 3. 연령대별 업종 선호도 시각화 (상위 5개씩)
fig, axes = plt.subplots(2, 3, figsize=(24, 16))
axes = axes.flatten()

for i, (col, label) in enumerate(zip(age_columns, age_labels)):
    ax = axes[i]
    top5 = age_industry_sales.sort_values(col, ascending=False).head(5)
    
    # 억 단위로 변환
    sales_data = top5[col] / 1e8
    
    sns.barplot(x=sales_data, y=top5['서비스_업종_코드_명'], ax=ax)
    ax.set_title(f'{label} 상위 5개 업종', fontsize=14, fontweight='bold')
    ax.set_xlabel('매출액 (억원)', fontsize=12)
    ax.set_ylabel('업종명', fontsize=12)
    
    # 수치 표시
    for j, v in enumerate(sales_data):
        ax.text(v + max(sales_data)*0.01, j, f'{v:.0f}억', 
                va='center', fontsize=10)

plt.tight_layout()
plt.show()

# 4. 업종별 연령대 구성비 분석 (히트맵)
# 각 업종에서 연령대별 매출 비율
age_industry_ratio = age_industry_sales.copy()
for col in age_columns:
    age_industry_ratio[col + '_비율'] = (age_industry_ratio[col] / 
                                       age_industry_ratio[age_columns].sum(axis=1) * 100)

# 상위 20개 업종만 선택
total_sales = age_industry_sales[age_columns].sum(axis=1)
top20_industries = age_industry_sales.loc[total_sales.sort_values(ascending=False).head(20).index]

# 비율 데이터 준비
ratio_columns = [col + '_비율' for col in age_columns]
ratio_data = age_industry_ratio.loc[top20_industries.index, ratio_columns]
ratio_data.columns = age_labels
ratio_data.index = top20_industries['서비스_업종_코드_명']

# 히트맵 그리기
plt.figure(figsize=(12, 16))
sns.heatmap(ratio_data, annot=True, fmt='.1f', cmap='YlOrRd', 
            cbar_kws={'label': '연령대별 매출 비율 (%)'})
plt.title('주요 업종별 연령대 매출 구성비 (%)', fontsize=16, fontweight='bold')
plt.xlabel('연령대', fontsize=12)
plt.ylabel('업종명', fontsize=12)
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# 5. 연령대별 특화 업종 분석 (각 연령대가 다른 연령대 대비 많이 쓰는 업종)
print("\n=== 연령대별 특화 업종 (상대적 선호도) ===")

# 전체 평균 대비 각 연령대의 특화 업종 찾기
for i, (col, label) in enumerate(zip(age_columns, age_labels)):
    # 해당 연령대 매출 비율 계산
    age_ratio = age_industry_sales[col] / age_industry_sales[col].sum() * 100
    
    # 전체 평균 대비 특화도 계산 (해당 연령대 비율 / 전체 평균 비율)
    total_avg = age_industry_sales[age_columns].sum(axis=1) / age_industry_sales[age_columns].sum().sum() * 100
    specialization = age_ratio / total_avg
    
    # 특화도 상위 10개 업종
    specialized_industries = specialization.sort_values(ascending=False).head(10)
    
    print(f"\n {label} 특화 업종 (전체 평균 대비):")
    for idx, spec_score in specialized_industries.items():
        industry_name = age_industry_sales.loc[idx, '서비스_업종_코드_명']
        print(f"  {industry_name}: {spec_score:.2f}배 특화")

# 6. 연령대 간 소비 패턴 유사성 분석
from sklearn.metrics.pairwise import cosine_similarity

# 연령대별 업종 소비 벡터
age_vectors = age_industry_sales[age_columns].T
similarity_matrix = cosine_similarity(age_vectors)

# 유사성 매트릭스 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(similarity_matrix, 
            xticklabels=age_labels, 
            yticklabels=age_labels,
            annot=True, fmt='.3f', cmap='coolwarm', center=0.8)
plt.title('연령대 간 소비 패턴 유사성', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n=== 연령대 간 소비 패턴 유사성 분석 ===")
for i in range(len(age_labels)):
    for j in range(i+1, len(age_labels)):
        similarity = similarity_matrix[i][j]
        print(f"{age_labels[i]} vs {age_labels[j]}: {similarity:.3f} (유사도)")
```

```python
# 3-3. 상위 3개 상권의 연령대별 매출 비교

top3_areas_df = df[df['상권_코드_명'].isin(top3_areas)]
age_by_area = pd.DataFrame()

for area in top3_areas:
    area_data = df[df['상권_코드_명'] == area]
    age_by_area[area] = area_data[age_cols].sum().values

age_by_area.index = ['10대', '20대', '30대', '40대', '50대', '60대 이상']

# 시각화: 상위 3개 상권의 연령대별 매출 비교
plt.figure(figsize=(12, 6))
age_by_area.plot(kind='bar')
plt.title('상위 3개 상권의 연령대별 매출 비교')
plt.xlabel('연령대')
plt.ylabel('매출액')
plt.legend(title='상권명')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

## 노량진: 60대 이상, 가산디지털단지: 40대, 용산전자상가: 50대 연령대에서 매출이 높게 나옴
```

### 4. 업종별 분석

```python
# 4-1. 업종별 총 매출액 및 매출 건수

category_count = df.groupby('서비스_업종_코드_명').size().reset_index(name='점포수')
category_sales = df.groupby('서비스_업종_코드_명').agg({
    '당월_매출_금액': 'sum',
    '당월_매출_건수': 'sum'
}).reset_index()


category_stats = pd.merge(category_sales, category_count, on='서비스_업종_코드_명')
print(category_stats)

# 4-2. 업종별 점포당 평균 매출액 및 객단가 계산

category_stats['점포당_평균_매출액'] = category_stats['당월_매출_금액'] / category_stats['점포수']
category_stats['객단가'] = category_stats['당월_매출_금액'] / category_stats['당월_매출_건수']

# 시각화 1: 점포당 평균 매출액 기준 상위 10개 업종
top_avg_categories = category_stats.sort_values('점포당_평균_매출액', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='점포당_평균_매출액', y='서비스_업종_코드_명', data=top_avg_categories)
plt.title('점포당 평균 매출액 기준 상위 10개 업종')
plt.xlabel('점포당 평균 매출액')
plt.ylabel('업종명')
plt.tight_layout()
plt.show()

# 시각화 2: 객단가 기준 상위 10개 업종
top_per_transaction = category_stats.sort_values('객단가', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='객단가', y='서비스_업종_코드_명', data=top_per_transaction)
plt.title('객단가 기준 상위 10개 업종')
plt.xlabel('객단가 (원/건)')
plt.ylabel('업종명')
plt.tight_layout()
plt.show()
```

##### 점포당 평균 매출액 기준 상위 업종은 수산물 판매, 컴퓨터 및 주변장치 판매, 가전제품, 청과상 등이 있다.
##### 객단가 기준 상위 업종은 의료기기, 시계 및 귀금속, 외국어학원, 가구 , 가전제품 등이 있다.

```python
# 4-3. 상위 3개 업종(점포당 매출액 기준)의 시간대별 매출 패턴

top3_categories = top_avg_categories['서비스_업종_코드_명'].head(3).tolist()
top_cat_df = df[df['서비스_업종_코드_명'].isin(top3_categories)]

time_by_category = pd.DataFrame()
for category in top3_categories:
    cat_data = top_cat_df[top_cat_df['서비스_업종_코드_명'] == category]

    category_store_count = category_count[category_count['서비스_업종_코드_명'] == category]['점포수'].values[0]
    time_by_category[category] = cat_data[time_cols].sum().values / category_store_count

time_by_category.index = ['00~06', '06~11', '11~14', '14~17', '17~21', '21~24']

# 시각화 3: 상위 3개 업종의 점포당 시간대별 매출 패턴
plt.figure(figsize=(12, 6))
time_by_category.plot(kind='line', marker='o')
plt.title('상위 3개 업종의 점포당 시간대별 매출 패턴')
plt.xlabel('시간대')
plt.ylabel('점포당 평균 매출액')
plt.legend(title='업종명')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 4-4. 점포수 대비 매출액 비율 

# 전체 점포수 대비 각 업종의 점포수 비율
category_stats['점포수_비율'] = category_stats['점포수'] / category_stats['점포수'].sum()
# 전체 매출액 대비 각 업종의 매출액 비율
category_stats['매출액_비율'] = category_stats['당월_매출_금액'] / category_stats['당월_매출_금액'].sum()
# 매출 효율성 = 매출액 비율 / 점포수 비율
category_stats['매출_효율성'] = category_stats['매출액_비율'] / category_stats['점포수_비율']


print(category_stats)
# 시각화 4: 매출 효율성 기준 상위 10개 업종
top_efficiency = category_stats.sort_values('매출_효율성', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='매출_효율성', y='서비스_업종_코드_명', data=top_efficiency)
plt.title('매출 효율성 기준 상위 10개 업종')
plt.xlabel('점포수 대비 매출 비율')
plt.ylabel('업종명')
plt.tight_layout()
plt.show()
```

```python
# 업종별 최적 행정동 추천 시스템

# 1. 업종별 행정동 분석 함수 
def analyze_best_dong_for_industry(df, target_industry, top_n=10):
    """
    특정 업종에 대해 최적의 행정동을 추천하는 함수
    """
    print(f"\n {target_industry} 업종 최적 입지 분석")
    print("="*50)
    
    # 해당 업종 데이터만 필터링
    industry_data = df[df['서비스_업종_코드_명'] == target_industry].copy()
    
    if len(industry_data) == 0:
        print(f" {target_industry} 데이터가 없습니다.")
        return None
    
    # 행정동별 집계
    dong_analysis = industry_data.groupby('행정동_코드_명').agg({
        '당월_매출_금액': 'sum',
        '당월_매출_건수': 'sum',
        '서비스_업종_코드_명': 'count'  # 해당 업종 점포 수
    }).reset_index()
    
    dong_analysis.columns = ['행정동', '총매출액', '총매출건수', '점포수']
    
    # 핵심 지표 계산
    dong_analysis['점포당_평균매출'] = dong_analysis['총매출액'] / dong_analysis['점포수']
    dong_analysis['객단가'] = dong_analysis['총매출액'] / dong_analysis['총매출건수']
    dong_analysis['점포당_매출건수'] = dong_analysis['총매출건수'] / dong_analysis['점포수']
    
    # 경쟁 강도 (점포 밀도) 계산
    dong_analysis['경쟁강도'] = pd.cut(dong_analysis['점포수'], 
                                     bins=5, labels=['매우낮음', '낮음', '보통', '높음', '매우높음'])
    
    # 종합 점수 계산 (정규화 후 가중평균)
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    
    dong_analysis['매출점수'] = scaler.fit_transform(dong_analysis[['점포당_평균매출']])
    dong_analysis['객단가점수'] = scaler.fit_transform(dong_analysis[['객단가']])
    dong_analysis['건수점수'] = scaler.fit_transform(dong_analysis[['점포당_매출건수']])
    
    dong_analysis['종합점수'] = (dong_analysis['매출점수'] * 0.6 + 
                               dong_analysis['객단가점수'] * 0.3 + 
                               dong_analysis['건수점수'] * 0.1)
    
    # 결과 정렬
    result = dong_analysis.sort_values('종합점수', ascending=False).head(top_n)
    
    return result

# 2. 어떤 업종들이 있는지 먼저 확인
print("📋 데이터에 있는 주요 업종들:")
industry_counts = df2['서비스_업종_코드_명'].value_counts().head(20)
for i, (industry, count) in enumerate(industry_counts.items(), 1):
    print(f"{i:2d}. {industry}: {count}개")

# 3. 주요 업종들에 대해 분석 실행
target_industries = ['한식음식점', '커피-음료', '미용실', '분식전문점', '슈퍼마켓']

all_results = {}

for industry in target_industries:
    result = analyze_best_dong_for_industry(df2, industry, top_n=15)
    if result is not None:
        all_results[industry] = result
        
        print(f"\n📊 {industry} 상위 10개 추천 행정동:")
        print("-" * 80)
        for idx, row in result.head(10).iterrows():
            rank = result.index.get_loc(idx) + 1
            print(f"{rank:2d}위. {row['행정동']:20s} "
                  f"매출:{row['점포당_평균매출']/1e8:.1f}억원 "
                  f"객단가:{row['객단가']:,.0f}원 "
                  f"점포수:{row['점포수']:2d}개 "
                  f"경쟁:{row['경쟁강도']} "
                  f"점수:{row['종합점수']:.3f}")

# 4. 시각화
if all_results:
    # 업종별 상위 5개 행정동 매출 비교
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    axes = axes.flatten()
    
    for i, (industry, data) in enumerate(all_results.items()):
        if i < 6:
            ax = axes[i]
            top5_data = data.head(5)
            
            # 점포당 평균매출을 억 단위로 변환
            sales_in_billion = top5_data['점포당_평균매출'] / 1e8
            
            bars = ax.bar(range(len(top5_data)), sales_in_billion, 
                         color=plt.cm.Set3(i), alpha=0.8)
            
            ax.set_title(f'{industry} 상위 5개 행정동', fontsize=12, fontweight='bold')
            ax.set_xlabel('행정동')
            ax.set_ylabel('점포당 평균매출 (억원)')
            ax.set_xticks(range(len(top5_data)))
            ax.set_xticklabels(top5_data['행정동'], rotation=45, ha='right')

            for j, bar in enumerate(bars):
                height = bar.get_height()
                ax.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                       f'{height:.1f}억', ha='center', va='bottom', fontsize=9)
    
    for i in range(len(all_results), 6):
        axes[i].set_visible(False)
    
    plt.tight_layout()
    plt.show()

# 5. 특정 업종 상세 분석 함수
def detailed_dong_analysis(df, industry_name):
    """특정 업종의 행정동별 상세 분석"""
    
    industry_data = df[df['서비스_업종_코드_명'] == industry_name].copy()
    
    if len(industry_data) == 0:
        print(f"{industry_name} 데이터가 없습니다.")
        return None
    
    # 행정동별 상세 분석
    detailed_analysis = industry_data.groupby('행정동_코드_명').agg({
        '당월_매출_금액': 'sum',
        '당월_매출_건수': 'sum',
        '주중_매출_금액': 'sum', 
        '주말_매출_금액': 'sum',
        '남성_매출_금액': 'sum',
        '여성_매출_금액': 'sum',
        '연령대_20_매출_금액': 'sum',
        '연령대_30_매출_금액': 'sum',
        '연령대_40_매출_금액': 'sum',
        '연령대_50_매출_금액': 'sum',
        '연령대_60_이상_매출_금액': 'sum',
        '서비스_업종_코드_명': 'count'
    }).reset_index()
    
    # 비율 계산
    detailed_analysis['객단가'] = detailed_analysis['당월_매출_금액'] / detailed_analysis['당월_매출_건수']
    detailed_analysis['주말_비율'] = detailed_analysis['주말_매출_금액'] / detailed_analysis['당월_매출_금액']
    detailed_analysis['여성_비율'] = detailed_analysis['여성_매출_금액'] / detailed_analysis['당월_매출_금액']
    detailed_analysis['점포당_매출'] = detailed_analysis['당월_매출_금액'] / detailed_analysis['서비스_업종_코드_명']
    
    # 주요 연령대 찾기
    age_cols = ['연령대_20_매출_금액', '연령대_30_매출_금액', '연령대_40_매출_금액', 
                '연령대_50_매출_금액', '연령대_60_이상_매출_금액']
    detailed_analysis['주요_연령대'] = detailed_analysis[age_cols].idxmax(axis=1).str.replace('_매출_금액', '').str.replace('연령대_', '')
    
    # 상위 10개 정렬
    top_dongs = detailed_analysis.sort_values('점포당_매출', ascending=False).head(10)
    
    print(f"\n {industry_name} 행정동별 상세 분석 (상위 10개)")
    print("="*100)
    print(f"{'순위':<4} {'행정동':<15} {'점포당매출':<10} {'객단가':<8} {'점포수':<6} {'주말비율':<8} {'여성비율':<8} {'주요연령대':<8}")
    print("-" * 100)
    
    for idx, row in top_dongs.iterrows():
        rank = top_dongs.index.get_loc(idx) + 1
        print(f"{rank:<4} {row['행정동_코드_명']:<15} {row['점포당_매출']/1e8:<10.1f} "
              f"{row['객단가']:<8,.0f} {row['서비스_업종_코드_명']:<6} "
              f"{row['주말_비율']:<8.1%} {row['여성_비율']:<8.1%} {row['주요_연령대']:<8}")
    
    return top_dongs

# 6. 한식음식점 상세 분석 실행
detailed_dong_analysis(df2, '한식음식점')
```
